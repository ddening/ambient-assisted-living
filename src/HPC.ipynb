```python
from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_dir = "../data/amd_reference_data.csv"
df_reference = pd.read_csv(
    data_dir
)  # names=('method', 'threads', 'frames', 'ssize', 't_mean')
df_reference = df_reference.sort_values(by="threads")
display(df_reference)
# print(df_reference.to_string())

method = df_reference["method"][0]
df_reference.plot(
    x="threads", y="time_mean", style="rx--", label=method
)  # kind='scatter'

plt.legend(loc="best")
plt.title("Parallel-Reference MPI Runtime")
plt.xlabel("Threads")
plt.ylabel("Time in s")
plt.grid()
plt.show()

df_reference.corr(method="pearson", numeric_only=True)

data = {
    "threads": df_reference["threads"],
    "time_mean": df_reference["time_mean"],
}

df = pd.DataFrame(data)
df.corr(method="pearson")
```
